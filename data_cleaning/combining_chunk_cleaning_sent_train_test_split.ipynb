```python
import re
import numpy as np

nepali_char_range = r'\u0900-\u097F'
allowed_punctuation = '।'

# Function to clean a sentence
def clean_sentence(sentence):
    # Remove all characters that are not Nepali or allowed punctuation
    cleaned = re.sub(f'[^{nepali_char_range}{allowed_punctuation} ]', '', sentence)
    # Remove extra spaces
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()
    return cleaned

# Apply cleaning to
```
